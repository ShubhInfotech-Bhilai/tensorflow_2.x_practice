```python
import matplotlib.pyplot as plt
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
np.random.seed(1)
tf.random.set_seed(1)
```

```python
class LinearRegressionModel(tf.keras.Model):
    
    def __init__(self):
        super().__init__()
        self.L1 = tf.keras.layers.Dense(units=1)
        
    def call(self, inputs):
        self.x = self.L1(inputs)
        return self.x
```

```python
model = LinearRegressionModel()
```

```python
data = pd.read_csv('dataset/linear_regression.txt').to_numpy()
```

```python
x = data[:, 0].reshape(data.shape[0], 1)
y = data[:, 1].reshape(data.shape[0], 1)

print('X Shape:', x.shape)
print('Y shape:', y.shape)
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
hist = model.fit(x=x, y=y, epochs=5000, verbose=0)
```

```python
plt.plot(hist.history['loss'])
plt.show()
```

```python
x_p = np.linspace(5, 25, num=20).reshape(20, 1)
y_p = model.predict(x_p)

plt.scatter(x, y)
plt.plot(x_p, y_p, c='r')
plt.show()
```
